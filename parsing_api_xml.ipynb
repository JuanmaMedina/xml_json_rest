```python
## Parsing API XML response data from PRIDE ##

import requests
import xml.etree.ElementTree as ET

# API call
r = requests.get('http://proteomecentral.proteomexchange.org/cgi/GetDataset?ID=9961&outputMode=XML&test=no')

# Parse data into a dictionary
root = ET.fromstring(r.content)

# View all the tags in the XML
for child in root.iter('*'):
    print(child.tag)

# Access atribute information from specific tag: atribute-metadata as key-value pairs
# for child in root.iter('cvParam'):
#     print(child.tag, child.attrib)

# Access specific atribute information from specific tag
# for child in root.iter('cvParam'):
#     print(child.attrib['cvRef'], child.attrib['accession'])

# Example: retrieve all the accession numbers in the XML
acc_list = []

for child in root.iter('cvParam'):
    acc_list.append(child.attrib['accession'])
    
print('Total accession numbers:', acc_list)
```
